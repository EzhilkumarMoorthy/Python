```python
# __init__ method will be called whenever the new objet is created. It's like a constructor
# First argument is 'self' for all methods which is created inside of the class
# All variables created inside the class are static

class myclass:
        sample = 0
        w = 0
        b = 0
        g = 0
        o = 0
        color_list = ['black','white','gray']
        
        def __init__(self, name):
            myclass.name = name
            print("What's the color of your car??")
            myclass.color = input()
            myclass.sample = myclass.sample + 1
        
        def check_color(self):
            if(myclass.color in myclass.color_list):
                if(myclass.color == myclass.color_list[0]):
                    myclass.b = myclass.b + 1
                elif(myclass.color == myclass.color_list[1]):
                    myclass.w = myclass.w + 1
                else:
                    myclass.g = myclass.g + 1
            else:
                myclass.o = myclass.o + 1
        
        def display_result(self):
            print("Hello!! Name is ", myclass.name)
            print("Total no.of black cars ", myclass.b)
            print("Total no.of white cars ", myclass.w)
            print("Total no.of gray cars ", myclass.g)
            print("Other color cars ", myclass.o)
            print("Sample size ",myclass.sample)
            return

"""
myobj1 = myclass("Ezhil")
myobj1.check_color()
myobj1.display_result()
            
myobj2 = myclass("Manju")
myobj2.check_color()
myobj2.display_result()            
"""
mylist = []
cnt = 0
while(cnt < 4):
    print("Enter your name.")
    actual_name = input()
    mylist.append(myclass(actual_name))
    mylist[cnt].check_color()
    mylist[cnt].display_result()
    cnt = cnt + 1
```
